```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/TemperatureofIndia.csv")
df.head()
```

```python
df['ANNUAL'].max()
```

```python
df['ANNUAL'].min()
```

```python
column_list = ['ANNUAL', 'JAN', 'FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
df[column_list].describe()
```

```python

plt.figure(figsize=(15,5), dpi = 500)
plt.title("TEMPERATURE OF INDIA FROM 1901 to 2017", fontdict={'color': 'red'})


plt.xlabel("YEARS from 1901 to 2017",fontdict={'color': 'brown'})
plt.ylabel("TEMPERATURES in *C",fontdict={'color': 'brown'})
plt.scatter(x=df['YEAR'][:51], y=df['ANNUAL'][:51], c='blue', label='First 50 Years', alpha=0.7,s=50)

plt.scatter(x=df['YEAR'][51:], y=df['ANNUAL'][51:], c='green', label='Remaining Years', alpha=0.7,s=50)

plt.xticks([1901, 1911, 1921, 1931, 1941, 1951, 1961, 1971, 1981, 1991, 2001, 2011, 2021])
plt.yticks([0, 10, 20, 30, 40, 50])

plt.legend()
plt.show()

```

```python
correlation = df.corr()
correlation.head()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.2f', annot=True, annot_kws={'size': 8}, cmap='magma')
```
